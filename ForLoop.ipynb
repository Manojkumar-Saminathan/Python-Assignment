```python
lists=[10,23,41,53,68]
for age in lists:
    if(age<18):
        print('age',age,'children')
    elif(age<45):
        print('age',age,'Adult')
    elif(age<60):
        print('age',age,'citizen')
    else:
        print('age',age,'Senior Citizen')
```
