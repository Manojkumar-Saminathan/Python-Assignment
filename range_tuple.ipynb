```python
value=['Name','Age','Profession']
for temp in range(len(value)):
    if value[temp]=='Name':
        print('-',input('Enter Your Name'),'-')
    elif value[temp]=='Age':
        print('-',int(input('Enter Your Age')),'-')
    else:
        print('-',input('Enter Your Profession'),'-')
```
